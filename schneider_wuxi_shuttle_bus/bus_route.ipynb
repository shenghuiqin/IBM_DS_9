# schneider无锡班车线路图
参考数据：https://github.com/shenghuiqin/IBM_DS_9/tree/master/schneider_wuxi_shuttle_bus

```python
import folium
import pandas as pd
df = pd.read_excel("F:\OneDrive\Job\Schneider\班车线路.xlsx")

# create wuxi map and display it
wx_map=folium.Map(location=[31.536164, 120.335941],
            zoom_start=11,
            #高德底图
            tiles='http://webrd02.is.autonavi.com/appmaptile?lang=zh_cn&size=1&scale=1&style=7&x={x}&y={y}&z={z}',
            attr='default'
        )

# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the station and add each to the incidents feature group
# route 1 blue circle
for lat, lng, in zip(df.lat, df.lng):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
wx_map.add_child(incidents)

# loop through the points and add each to the map
# route 2 green circle
for lat, lng, in zip(df.lat2, df.lng2):
    folium.CircleMarker(
        [lat, lng],
        radius=5, # define how big you want the circle markers to be
        color='darkred',
        fill=True,
        fill_color='green',
        fill_opacity=0.6
    ).add_to(wx_map)

# route 3 pink circle
for lat, lng, in zip(df.lat3, df.lng3):
    folium.CircleMarker(
        [lat, lng],
        radius=5, # define how big you want the circle markers to be
        color='lightgray',
        fill=True,
        fill_color='black',
        fill_opacity=0.6
    ).add_to(wx_map)
# add company location 
folium.Marker([31.53733388, 120.3711032	], popup="Schneider", icon=folium.Icon(icon='building',prefix='fa')).add_to(wx_map)
# add unsure location 
folium.Marker([31.528749, 120.504504], popup="shuofang taikang not found", icon=folium.Icon(icon="question",prefix='fa')).add_to(wx_map)
wx_map
```
